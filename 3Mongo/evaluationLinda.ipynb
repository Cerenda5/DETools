```python
import pandas as pd
import pymongo
```

```python
client = pymongo.MongoClient()
database = client['exercices']
collection = database['sanctions']
```

```python
df_sanctions = pd.read_csv("./data/sanctionspolice.csv")
df_sanctions.head()
```

## Netoyage des données

```python
l_d = df_sanctions.to_dict(orient='records')
```

## Importer les données

```python
from pymongo import MongoClient
client = MongoClient("mongo:27017")
```

```python
df_sanctions = client.ks
collection_linda_sanctions = df_sanctions['lindadb']
```

```python
collection_linda_sanctions.delete_many({})
collection_linda_sanctions.insert_many(l_d)
```

## Requête 1

```python
req = collection_linda_sanctions.find().limit(3).sort([("nombre", -1)])
for test in req:
    print(test)
```

## Requête 2

```python
req = collection_linda_sanctions.find({"sanction":"Blâme"})
list(req)
```

## Requête 3

```python
req = collection_linda_sanctions.aggregate([{ "$group": { "_id": "$sanction"}  },{ "$group": { "_id": "sanction", "count": { "$sum": 1 } } } ])
list(req) 
```
